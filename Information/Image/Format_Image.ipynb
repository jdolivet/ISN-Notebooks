* ## Représentation matricielle

```python
def pbm_vers_matrice(fichier_texte):
    """ Retourne la représentation matricielle d'une image
    paramètre : fichier image au format PBM ASCII
    Imprime les dimensions de l'image"""
    # On récupère une liste de lignes à partir du fichier
    with open(fichier_texte, 'r') as source:
        liste_lignes = source.readlines()
    # On vérifie le format
    magic = liste_lignes[0].strip()
    if magic != "P1":
        return "L'image n'est pas au format attendu"
    # On récupère les informations de l'image
    dimensions = liste_lignes[2].split()
    largeur = int(dimensions[0])
    print("Largeur de l'image :", largeur)
    hauteur = int(dimensions[1])
    print("Hauteur de l'image :", hauteur)
    # On construit la matrice
    ligne = 3
    matrice = [[0 for j in range(largeur)] for i in range(hauteur)]
    for j in range(hauteur):
        for i in range(largeur):
            pixel = liste_lignes[ligne][i]
            matrice[j][i] = int(pixel)
        ligne += 1
    return matrice
```

```python
fichier = "Images/Joconde/Joconde.pbm"
matricePBM = pbm_vers_matrice(fichier)
# print(matricePBM[0])
```

```python
# On imprime le tableau de pixels
hauteur = len(matricePBM)
largeur = len(matricePBM[0])
for j in range(hauteur):
    for i in range(largeur):
        pixel = chr(matricePBM[j][i])
        print(pixel, end='')
    print()
```

## Représentation matricielle

```python
def ppm_vers_matrice(fichier_texte):
    # On prélève les informations nécessaires
    dimensions = fichier_texte[2].split()
    largeur = int(dimensions[0])
    print("Largeur de l'image :", largeur)
    hauteur = int(dimensions[1])
    print("Hauteur de l'image :", hauteur)
    nuances = int(fichier_texte[3])
    print("Nuances de couleurs :", nuances)

    # On construit la matrice
    ligne = 4
    matrice = [[0 for j in range(largeur)] for i in range(hauteur)]
    for j in range(hauteur):
        for i in range(largeur):
            pixel = [0,0,0]
            for k in range(3):
                pixel[k] = int(fichier_texte[ligne])
                ligne += 1
            matrice[j][i] = pixel
    return matrice
```

```python
# On importe l'image
from urllib.request import urlopen
source = urlopen("https://sites.google.com/site/sciencesdunumerique/information/images/Joconde.ppm")
image = source.readlines()

# On convertit en matrice
matricePPM = ppm_vers_matrice(image)
```

```python
# On imprime le tableau de pixels
def imprime_matrice_RGB(matrice):
    hauteur = len(matricePPM)
    largeur = len(matricePPM[0])
    for j in range(hauteur):
        for i in range(largeur):
            pixel = matricePPM[j][i]
            l = len(str(pixel))
            print(pixel, end=' '*(16 - l))
        print()
        
imprime_matrice_RGB(matricePPM)
```

# Récupérer matrice bitmap

Cette portion de code permet de récupérer la matrice de pixels d'autres formats grâce à des bibliothèques adaptées.

```python
%matplotlib inline
from matplotlib.pyplot import imshow
from PIL import Image
from io import BytesIO
import numpy as np
import requests

def image_depuis_internet(img_url):
    # Récupère l'image
    response = requests.get(img_url)
    img = Image.open(BytesIO(response.content))
    # Récupère la matrice bitmap
    bitmap = np.array(img)
    # Visualisation de l'image
    imshow(img)
    # Retourne la matrice bitmap
    return bitmap
```

```python
# Format jpg
url_jpg = "https://upload.wikimedia.org/wikipedia/commons/9/99/ClaudeShannon_MFO3807.jpg"
matriceJPG = image_depuis_internet(url_jpg)
```

```python
# On imprime le tableau de pixels
imprime_matrice_RGB(matriceJPG)
```

```python
# Format PNG
url_png = "https://upload.wikimedia.org/wikipedia/commons/thumb/b/b7/Turing_Machine.png/256px-Turing_Machine.png"
matricePNG = image_depuis_internet(url_png)
```

```python
# Format GIF
url_gif = "https://upload.wikimedia.org/wikipedia/commons/1/1a/Carnabotnet_geovideo_lowres.gif"
matriceGIF = image_depuis_internet(url_gif)
```
